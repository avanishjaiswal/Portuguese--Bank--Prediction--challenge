```python
Y=X["y"]
X_A=X.drop("y",axis=1)
```

```python
X_array=X_A.values.reshape(41188, 20 )
Y_array=Y.values.reshape(41188,1)
```

```python
print(type(Y_array[0][0]))
print(type(X_array[0][0]))
```

```python
Y_array.ravel
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_array,Y_array,test_size=.30,random_state=130) 
```

```python
Y_test=Y_test.astype('int')
Y_train=Y_train.astype('int')
```

```python
#using KNeighborsClassifier

from sklearn.neighbors import KNeighborsClassifier
teacher1=KNeighborsClassifier()
learner1=teacher1.fit(X_train,Y_train)

Yp1=learner1.predict(X_test)
```

```python
#using GaussianNB

from sklearn.naive_bayes import GaussianNB
teacher2=GaussianNB()
learner2=teacher2.fit(X_train,Y_train)

Yp2=learner2.predict(X_test)    
```

```python
#using DecisionTreeClassifier

from sklearn import tree
teacher3=tree.DecisionTreeClassifier()
learner3=teacher3.fit(X_train,Y_train)

Yp3=learner3.predict(X_test)    
```

```python
#using RandomForestClassifier

from sklearn.ensemble import RandomForestClassifier
teacher0 = RandomForestClassifier(n_estimators=250,random_state=0)
learner0=teacher0.fit(X_train,Y_train)
Yp0=learner0.predict(X_test)  
```

```python
from sklearn.metrics import accuracy_score

print("accuracy_score for RandomForestClassifier=",accuracy_score(learner0.predict(X_test),Y_test))
print("accuracy_score for DecisionTreeClassifier=",accuracy_score(learner3.predict(X_test),Y_test))
print("accuracy_score for GaussianNB=",accuracy_score(learner2.predict(X_test),Y_test))
print("accuracy_score for KNeighborsClassifier=",accuracy_score(learner1.predict(X_test),Y_test))
```

```python
#clearly we have highest accuracy_score for RandomForestClassifier and KNeighborsClassifier
```
